```python
df= df.dropna(how="any")
df = df.rename(columns= {'Disease': 'STD'})
df['STD Cases v Population Ratio'] = df['STD Cases'] / df['Population']
df[['STD Cases v Population Ratio', 'State']].describe
```

```python
df= df.drop(['index','Disease Code', 'Age'], axis =1)
df.head(10)
```

```python
df.sort_values(by = 'STD Cases v Population Ratio', ascending=False)
```

```python
ny_df = df[df['State'] == 'New York']
ny_df.sort_values(by = 'STD Cases v Population Ratio', ascending=False)
```

```python
df['Gender'] = df['Gender'].str.upper()
df.head
```

```python
numerical_present = not df.select_dtypes(include=['number']).empty
print(numerical_present)
```

```python
df.groupby('Age Code').agg(['mean', 'min', 'max'])
```

```python
state_stats = df.groupby('State')['Population'].agg(['mean', 'min', 'max'])
print(state_stats)
```
