```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

# Asset Allocation

```python
df = pd.read_csv('https://raw.githubusercontent.com/kglan/MSDS/main/DATA618/portfolio.csv')
```

```python
df = df[:-3]
df.head(10)
```

# Asset Allocation Visualizations

```python
# Remove the currency symbol and convert 'Amount' to numeric
df['Amount'] = df['Amount'].str.replace('$', '').str.replace(',', '').astype(float)

```

```python

# Group the data by 'Subtype' and sum the 'Amount' for each subtype
subtype_amount = df.groupby('Subtype')['Amount'].sum()
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(subtype_amount, labels=subtype_amount.index, autopct='%1.1f%%', startangle=140)
plt.title('Portfolio Allocation by Subtype')
plt.show()
```

```python
# Portfolio Allocation Pie Chart
allocation = df.groupby('Type')['Amount'].sum()
plt.figure(figsize=(8, 8))
plt.pie(allocation, labels=allocation.index, autopct='%1.1f%%', startangle=140)
plt.title('Portfolio Allocation')
plt.show()
```

```python
colors = ['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightseagreen', 'lightpink', 'lightsteelblue', 'lightyellow']
plt.figure(figsize=(12, 6))
bars = plt.bar(df['Ticker'], df['Amount'], color=colors)
plt.xlabel('Ticker')
plt.ylabel('Amount')
plt.title('Amount by Ticker')
plt.xticks(rotation=45)
# Customize colors for each bar
for i in range(len(bars)):
    bars[i].set_color(colors[i % len(colors)])  # Repeating colors if more bars than colors
plt.show()
```

```python
ticker_amount = df.groupby('Ticker')['Amount'].sum()
plt.figure(figsize=(8, 8))
plt.pie(ticker_amount, labels=ticker_amount.index, autopct='%1.1f%%', startangle=140)
plt.title('Amount Distribution by Ticker')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```
