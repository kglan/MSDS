```python
import networkx as nx
import matplotlib.pyplot as plt

x = {'LilKim': ['Notorious BIG', 'PuffDaddy', 'JayZ', 'Rihanna'], 
     'Notorious BIG': ['LilKim', 'JayZ', 'Beyonce', 'Kanye'], 
     'PuffDaddy': ['LilKim', 'JayZ', 'Rihanna'], 
     'JayZ': ['LilKim', 'Notorious BIG', 'PuffDaddy', 'Beyonce', 'Rihanna', 'Kanye'], 
     'Beyonce': ['Notorious BIG', 'JayZ', 'Kanye'], 
     'Rihanna': ['LilKim', 'PuffDaddy', 'JayZ', 'Kanye', 'SZA'], 
     'Kanye': ['Notorious BIG', 'JayZ', 'Beyonce', 'Rihanna', 'SZA'], 
     'SZA': ['MegTheeStallion', 'Rihanna', 'Kanye'],
     'MegTheeStallion': ['Usher', 'SZA'], 
     'Usher': ['MegTheeStallion']}

# Create a new graph object
G = nx.Graph()

# Add nodes and edges to the graph
for node in x.keys():
    G.add_node(node)
    for edge in x[node]:
        G.add_edge(node, edge)

# Draw the graph
plt.figure()
nx.draw(G, with_labels=True, ax=plt.gca())
plt.show()
```
